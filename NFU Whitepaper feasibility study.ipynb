# Feasibility study of a multi-modal Parkinson monitor system

#### Introduction

Parkinson’s disease (PD) is a neurodegenerative disorder characterized by motor symptoms, like tremor, rigidity, and bradykinesia, which fluctuate during the day. Nowadays, symptoms and disease course are charted by a clinician and are measured most often using the UPDRS. This is however associated with limitations: First, symptoms and disease course are assessed once every couple of weeks or months and therefore only permit a snapshot of the clinical situation. Second, scores given by the clinician have a highly subjective character. Third, patients might put themselves in a better light during in-clinical assessments compared to when they are at home. To get more frequent and more objective ratings of symptoms and disease course, continuous monitoring systems are needed. Such systems should represent the symptomatology in the daily life of patients and could eventually also be used as input signal for adaptive, responsive, or closed-loop deep brain stimulation (Habets et al., 2018, in production). Recently, a couple of monitoring systems have been developed and tested, and have shown promising results (Cancela, 2016; Angeles, 2017; Bayes, 2018). In this study, the feasibility of a unique and newly developed PD monitoring system, which combines wearables with an experience sampling method (ESM) application, was tested.

Wearables are increasingly used to detect PD symptoms due to the technological progress made (Sanchez-Ferro, 2016; Rovini, 2017). They have shown promise in detecting tremor (Basu et al., 2013; Khobragade et al., 2015; Delrobaei et al., 2018), freezing of gait (Rodriguez-Martin et al., 2017), bradykinesia, and dyskinesia (Griffiths, 2012; Delrobaei 2017). Wearables are included in the PD monitoring system to detect PD symptoms and the associated fluctuations in symptoms. 
The monitoring system also included an ESM smartphone application. ESM is a validated, digital diary method with multiple repeated measurements at semi-random moments in daily life (Csikszentmihalyi, 1987; Myin-Germes, 2009). It is beneficial compared to diary and cross-sectional assessments, because there is no recall bias and data is collected on multiple moments a day (Palmier-Claus, 2011). The designed ESM application included questions regarding mood states, contextual information and both motor and non-motor PD symptoms. By combining wearables with ESM, we aim to retrieve both objective and subjective information about the frequency and severity of PD motor and non-motor symptom fluctuations.

The feasibility of wearable sensors in PD is already proven since they are used widely in PD monitoring systems (Sanchez-Ferro, 2016; Rovini, 2017). The feasibility of an ESM application was demonstrated among 5 PD patients during 5 days (Broen, 2016), and among one PD patient during 34 days (van der Velden et al., 2018). The feasibility of these methods combined has been tested before (Booij et al., 2015; Brannon et al., 2016; Maher et al., 2016; Bedard et al., 2017; Knell et al., 2017; Liao et al., 2017), but not in PD patients. This project therefore aims to prove feasibility of our monitoring system in PD patients during daily life in two consecutive weeks.

#### Methods

##### Participants

This study was approved by the METC azM/UM and written informed consent was obtained from all participants. XXX idiopathic PD patients participated in this study. Recruitment was done through their neurologists or neurosurgeons at the Maastricht University Medical Centre. Only patients with idiopathic PD were included. Patients’ disease severity had to be rated as mild to severe (Hoehn and Yahr 1-4). Further, patients were included if they had an age between 18 and 80 years, if they were in possession of a smartphone (minimal iOS 8 or android 4), if they could fluently speak and read Dutch, if they were available during two consecutive weeks of representative daily activities (meaning no holidays or planned hospital admission), and if they gave written informed consent. Subjects were tested for cognitive deficits and were excluded if they scored less than 24 points on the Montreal Cognitive Assessment. Participant characteristics are shown in Table XXX.

#####  Procedure

When patients showed interest in the project, after their clinician informed them briefly about it, they were informed in more detail by the researchers and got an information folder to take home. When patients were willing to participate in this study, they had to come to the hospital twice; one start session preceding the measurement period of two weeks and one end session after the measurement period. During the start session the participants received explanation about the wearables and ESM application and they received the required materials. At the end session, participants completed an evaluation form and returned the materials.

##### Wearables

Wearables were developed by the Instrument Development Engineering & Evaluation department of the Maastricht University. Those wearables contained a 6 DOF sensor, consisting of an accelerometer and gyroscope. Self-developed wearables were used because of the possibilities to implement symptom algorithms and online data streaming for purposes out the scope of this study.
During the measurement period, each participant wore three wearables; two at each wrist and one at the chest. The wearables were attached to the body via handmade wristbands and a necklace (Figure 1). The participants wore the wearables in the daytime, and charged them at night. The participants were asked to remove the wearables when they were for example taking a shower, because although the wearables are water-resistant, the accessories would stay wet for a while which might be experienced as uncomfortable. Each participant used one charger block with three cables running from it. By first removing the charger block out of the outlet, it was made sure that the starting time of the three wearables were synchronous.

##### ESM

The ESM smartphone application was developed by Sananet, specialist in eHealth (or an ESM app, the Psymate app, was downloaded and installed on the smartphone of the participants.). Development of the used questionnaires is explained in more detail elsewhere (artikel Jeroen over vragenlijst). In brief, the questionnaires were developed using previous work (REF) and patient and clinician interviews about what is considered as important when identifying intraday fluctuations in PD symptoms.
During the measurement period, participants were asked to complete a morning (5 questions) and evening questionnaire (10 questions) which were identical on all days. In addition, they received seven repetitive questionnaires (27 questions) randomly during the day. The random questionnaires stayed available for 15 minutes and the participants were asked to complete as much questionnaires as possible without adapting their normal daily behaviour. Participants had to rate the questions on a Likert scale (1-7). See Figure X for a screenshot of the ESM application. See supplementary materials for morning, random and evening questionnaires in both Dutch and English.

```python
def measurementtime(filename):  
    f = pyedflib.EdfReader(filename) 
        #reading in edf file

    sr = f.getSampleFrequencies()[0] 
        #get sample freq out of edf file via edflib
    n = f.signals_in_file 
        #defining number of channels

    startingTime=filename[-19:-4] 
        #extract starting time from file name by specifying its index

    startingTime=pd.to_datetime(startingTime, format='%Y%m%d_%H%M%S', errors='ignore') 

    timeStamps=[] 
        #empty list here we are creating time of all measurements

    last=startingTime 

    for time in range(f.getNSamples()[0]): 
        #range from 0 till number of samples
        timeStamps.append(last)
        #startingTime is added as first value into timeStamp list
        last=last+pd.Timedelta('%d ms' %(1000/sr)) 
        #now the time of one sample is added to the starting time to create list
        #timeStamps consist of date (Y,M,D) and time (H,M,S,MS)
        
    measurementtime = timeStamps[-1] - timeStamps[0]
    return measurementtime
```
